```python
import pandas as pd
```

```python
df1 = pd.read_csv('./data/tmdb_5000_credits.csv')
df2 = pd.read_csv('./data/tmdb_5000_movies.csv')
```

```python
df1.head(1)
```

```python
df2.head(1)
```

```python
# 칼럼 이름 바꾸기

df1.columns = ['id', 'title', 'cast', 'crew']
df1.head(1)
```

```python
# 필요한 데이터 추출

temp = df1[['id', 'cast', 'crew']]
temp
```

```python
# 두개의 데이터 합치기

df2 = df2.merge(temp, on='id')
df2.head(1)
```

```python
df2['overview']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
```

```python
df2['overview'].isnull().values.any()
```

```python
df2['overview'] = df2['overview'].fillna('')
```

```python
df2['overview'].isnull().values.any()
```

```python
matrix = tfidf.fit_transform(df2['overview'])
```

```python
matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
sim = linear_kernel(matrix, matrix)
```

```python
sim.shape
```

```python
sim
```

```python
import pickle
```

```python
# 데이터 파일로 만들기

pickle.dump(sim, open('./data/sim.pickle', 'wb'))
```

```python
# 파일로 만든 데이터 가져오기

sim = pickle.load(open('./data/sim.pickle', 'rb'))
```

```python
sim
```

```python
movies = pickle.load(open('./data/movies.pickle', 'rb'))
```

```python
movies
```

```python
# 1. 제목에 대한 인덱스값 구한다.

filter = movies['title'] == 'The Dark Knight Rises'
idx = movies[filter].index[0]
```

```python
# 2. 유사도 테이블에서 인덱스번째 유사도 데이터를 구한다
sim_scores = sim[idx]
sim_scores
```

```python
# 3. 인덱스번호, 유사도 값을 배열로 생성

sim_scores = list(enumerate(sim_scores))
sim_scores
```

```python
# 4. 비교했을때 유사도가 가장 높은순으로 정렬

sim_scores = sorted(sim_scores, key=lambda x:x[1], reverse=True)
sim_scores
```

```python
# 5. 상위 10개의 값을 구한다.

sim_scores = sim_scores[1:13]
sim_scores
```

```python
# 6. 인덱스 배열 생성

sim_movies = [i[0] for i in sim_scores]
sim_movies
```

```python
for i in range(12):
    title = movies['title'].iloc[sim_movies[i]]
    print(title)
```
