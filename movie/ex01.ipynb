```python
df1 = pd.read_csv('./data/tmdb_5000_credits.csv')
df2 = pd.read_csv('./data/tmdb_5000_movies.csv')
```

```python
df1.columns=['id', 'title', 'cast', 'crew']
```

```python
df1.head(1)
```

```python
df_temp = df1[['id', 'cast', 'crew']]
df_temp.head()
```

```python
df2 = df2.merge(df_temp, on = 'id')
df2.head(1)
```

```python
# 점수평균

C = df2['vote_average'].mean()
C
```

```python
# 상위 10%평가수

m = df2['vote_count'].quantile(0.9)
m
```

```python
# 상위 10% 영화 수

filter = df2['vote_count'] >= m
q_movies = df2.loc[filter]
q_movies.shape
```

```python
# 리뷰가 적은순으로 출력

q_movies['vote_count'].sort_values()
```

```python
def weighted_rating(x, m = m, C = C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v + m) * R) + (m /(m + v) * C)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies.head(1)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)
```

```python
q_movies[['id', 'title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
q_movies = q_movies[['id', 'title', 'vote_count', 'vote_average', 'score']]
q_movies
```
